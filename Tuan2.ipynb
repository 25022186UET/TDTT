```python
#W2A3
a = int(input("Nhap so a: "))
b = int(input("Nhap so b: "))

print(f"Tong cua {a} va {b} la: {a + b}")
print(f"Hieu cua {a} va {b} la: {a - b}")
print(f"Tich cua {a} va {b} la: {a * b}")
print(f"Phan nguyen cua {a} va {b} la: {a // b}")
print(f"Phan du cua {a} va {b} la: {a % b}")
print(f"Ket qua chia thuc cua {a} va {b} la: {a / b:.2f}")
```

```python
#W2A4
a1, b1, c1, a2, b2, a3 = map(int, input("Nhap 6 dau diem: ").split())
TB = ((a1 + b1 + c1) + (a2 + b2) * 2 + a3 * 3) / 10
print(f"Diem trung binh la: {TB:.1f}")
```
