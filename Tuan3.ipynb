1. Nhap vao 1 so nguyen duong, kiem tra xem no co phai la nam nhuan hay khong (Nam chia het cho 4 va khong chia het cho 100)

```python
nam = int(input())
if (nam%4 == 0 and nam%100 !=0) or (nam%400 == 0):
    print(nam, "la nam nhuan")
else:
    print(nam, "khong phai la nam nhuan")
```

1. Nhung cach dung khac cua ham if-else

```python
n = int(input())
if (n>0):
    print(n, "la so duong")
elif (n<0):
    print(n, "la so am")
else:
    print(n, "la so khong")
```

1. Nhap vao 1 so nguyen duong n (0...9), doc so tuong ung

```python
n = int(input())
if 0 <= n <= 9:
    if n == 0:
        s = "So khong"
    elif n == 1:
        s = "So mot"
    elif n == 2:    
        s = "So hai"
    elif n == 3:    
        s = "So ba"
    elif n == 4:
        s = "So bon"
    elif n == 5:    
        s = "So nam"
    elif n == 6:    
        s = "So sau"    
    elif n == 7:
        s = "So bay"
    elif n == 8:    
        s = "So tam"
    else:    
        s = "So chin"
    print(s)
else:
   print("So ban nhap vao toi chua doc duoc") 
```

```python
n = int(input())
s = ""
match n:
    case 0:
        s = "So khong"
    case 1:
        s = "So mot"
    case 2:
        s = "So hai"
    case _:
        s = "skibidi"
print(s)
```

1. Nhap vao 1 so nguyen duong co 3 chu so, tim va in ra so dao nguoc

```python
n = int(input())
if 99 < n < 1000:
    dv = n % 10
    chuc = (n // 10) % 10
    tram = n // 100
    n_dao = dv*100 + chuc*10 + tram
    print(n_dao)
else:
    print("Chua xu ly duoc so nay")
```

1. Cho 1 so n, dai dien cho 1 thang trong nam (nam nhuan)
2. In ra so ngay tuong ung

```python
thang = int(input())
match thang:
    case 1|3|5|7|8|10|12:
        s = f"Thang {thang} co 31 ngay"
    case 2:
        s = f"Thang {thang} co 29 ngay"
    case 4|6|9|11:
        s = f"Thang {thang} co 30 ngay"
    case _:
        s = "Khong co thang nay"
print(s)
```
