```python
#W4A1

#for
n = int(input())
tong = 0
for i in range(1, n+1):
    tong += i
print(tong)

#while
n = int(input())
tong = 0
i = 1
while i<=n:
    tong += i
    i += 1
print(tong)
```

```python
#W4A2

while True:
    n = int(input())
    if n>0:
        break
if n<2:
    print("Khong phai so nguyen to")
else:
    so_nguyen = True
    for i in range(2,n):
        if n%i==0:
            so_nguyen = False
            break
    if so_nguyen:
        print("La so nguyen to")
    else:
        print("Khong phai so nguyen to")
```

```python
#W4A3

while True:
    n = int(input())
    if 0<n<100:
        break
giai_thua = 1
for i in range(1, n+1):
    giai_thua *= i
print(giai_thua)
```

```python
#W4A6

while True:
    a, b = map(int, input().split())
    if a<=b:
        break
def is_prime(n):
    if n<2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n%i==0:
            return False
    return True
tong = 0
for i in range(a, b + 1):
    if is_prime(i):
        tong += i
print(tong)
```
