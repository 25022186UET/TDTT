# W4A1
Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để 
tính tống các số nguyên từ 1 đến n. (n <= 1000)

```python
#for
n = int(input())
tong = 0
for i in range(1, n + 1):
    tong += i
print(tong)

#while
n = int(input())
tong = 0
i = 1
while i <= n:
    tong += i
    i += 1
print(tong)
```

# W4A2
Viết chương trình nhập vào một số tới khi được một số nguyên dương thì thôi. 
Kiểm tra số này có phải là số nguyên tố hay không?

```python
while True:
    n = int(input())
    if n > 0:
        break
if n < 2:
    print("Khong phai so nguyen to")
else:
    so_nguyen = True
    for i in range(2, n):
        if n % i == 0:
            so_nguyen = False
            break
    if so_nguyen:
        print("La so nguyen to")
    else:
        print("Khong phai so nguyen to")
```

# W4A6
Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn 
[a, b]. Mỗi dòng là một bộ test (a b).

```python
while True:
    a, b = map(int, input().split())
    if a <= b:
        break
tong = 0
for i in range(a, b+1):
    nguyen_to = True
    if i < 2:
        nguyen_to = False
    else:
        for j in range(2, int(i**0.5)+1):
            if i % j == 0:
                nguyen_to = False
                break
    if nguyen_to:
        tong += i
print(tong)
```

# W4A7
Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n.

```python
while True:
    n = int(input())
    if n >= 2:
        break
i = 2
max_prime = -1
while i * i < n:
    while n % i == 0:
        max_prime = i
        n //= i
    i += 1
if n > 1:
    max_prime = n
print(max_prime)
```

# W4A22
Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ 

```python
n = int(input())
so_chan = 0
so_le = 0
while n > 0:
    so = n % 10
    if so % 2 == 0:
        so_chan += 1
    else:
        so_le += 1
    n = n // 10
print(so_chan, "so chan", so_le, "so le")
```

# W4A24
Nhập vào A, tìm n nhỏ nhất sao cho 
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 

```python
A = float(input())
n = 0
S = 0
while S <= A:
    n += 1
    S += 1/n
print(f"n nhỏ nhất là: {n}")
print(f"Tổng 1 + ... + 1/{n} = {S}")
```

# W4A25
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1. 
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
max_num = None
min_num = None
while True:
    num = int(input())
    if num == -1:
        break
    if max_num == None:
        max_num = num
        min_num = num
    else:
        if num > max_num:
            max_num = num
        if num < min_num:
            min_num = num
if max_num != None:
    print("Số lớn nhất:", max_num)
    print("Số nhỏ nhất:", min_num)
```

# W4A31
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in 
thường, bao nhiêu ký tự số

```python
chuoi = input()
in_hoa = 0
in_thuong = 0
so = 0
for ky_tu in chuoi:
    if ky_tu.isupper():
        in_hoa = in_hoa + 1
    elif ky_tu.islower():
        in_thuong = in_thuong + 1
    elif ky_tu.isdigit():
        so = so + 1
print("Ký tự in hoa:", in_hoa)
print("Ký tự in thường:", in_thuong)
print("Ký tự số:", so)
```
