# W5A1
Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. def Max_Of_Two( a, b) -> 
int: 

```python
def Max_Of_Two(a, b):
    if a > b:
        return a
    else:
        return b
a, b = map(int, input().split())
print(Max_Of_Two(a, b))
```

```python
def swap(a, b):
    return b, a
a, b = map(int, input().split())
a, b = swap(a, b)
print(a, b)
```

# W5A3
Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
def nguyen_to(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
n = int(input())
if nguyen_to(n):
    print("Yes")
else:
    print("No")
```

# W5A4
VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên.

```python
def perfect(n):
    if n < 2:
        return False
    tong_uoc = 1
    for i in range(2, n):
        if n % i == 0:
            tong_uoc += i
    return tong_uoc == n
n = int(input())
if perfect(n):
    print("Yes")
else:
    print("No")
```

# W5A5
Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1 

```python
def tim_vi_tri(day, n):
    if n in day:
        return day.index(n)
    else:
        return -1
day = list(map(int, input().split()))
n = int(input())
print(tim_vi_tri(day, n))
```

# W5A6
Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!) 

```python
def giai_thua(n):
    if n < 0:
        return None
    elif n == 0 or n == 1:
        return 1
    else:
        ket_qua = 1
        for i in range(2, n + 1):
            ket_qua *= i
        return ket_qua
n = int(input())
print(giai_thua(n))
```

# W5A7
Viết hàm giả lập một máy tính bỏ túi. 
Hàm thực hiện tính toán kết quả của phép toán giữa số \( num1 \) và \( num2 \) với toán tử \( operat 
\) 
Biết các toán tử đơn giản thực hiện phép toán lần lượt là: 
Phép trừ \( - \) 
Phép cộng \( + \) 
Phép chia \( / \) 
Phép nhân \( * \) 
Kết quả làm tròn đến chữ số thập phân thứ hai. 
Đầu vào: Hai số thực và phép toán cần thực hiện (trong số các phép toán \(+, -, *, /\)). 
Đầu ra: Kết quả phép toán. 

```python
def may_tinh(a, b, op):
    if op == "+":
        return a + b
    if op == "-":
        return a - b
    if op == "*":
        return a * b
    if op == "/":
        if b == 0:
            return None
        return a / b
a, b, op = input().split()
a = float(a)
b = float(b)
res = may_tinh(a, b, op)
if res == None:
    print("None")
else:
    print(f"{may_tinh(a, b, op):.2f}")
```

# W5A9
Viết hàm trả về tổng các chữ số của số nguyên dương n 

```python
def tong_so_nguyen(n):
    if n <= 0:
        return None
    tong = 0
    while n > 0:
        tong += n % 10
        n //= 10
    return tong
n = int(input())
print(tong_so_nguyen(n))
```
