# W5A3
Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
def nguyen_to(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
n = int(input())
if nguyen_to(n):
    print("Yes")
else:
    print("No")
```

# W5A4
VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên.

```python
def perfect(n):
    if n < 2:
        return False
    tong_uoc = 1
    for i in range(2, n):
        if n % i == 0:
            tong_uoc += i
    return tong_uoc == n
n = int(input())
if perfect(n):
    print("Yes")
else:
    print("No")
```

# W5A6
Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!) 

```python
def giai_thua(n):
    if n < 0:
        return None
    elif n == 0 or n == 1:
        return 1
    else:
        ket_qua = 1
        for i in range(2, n + 1):
            ket_qua *= i
        return ket_qua
n = int(input())
print(giai_thua(n))
```

# W5A7
Viết hàm giả lập một máy tính bỏ túi. 
Hàm thực hiện tính toán kết quả của phép toán giữa số \( num1 \) và \( num2 \) với toán tử \( operat 
\) 
Biết các toán tử đơn giản thực hiện phép toán lần lượt là: 
Phép trừ \( - \) 
Phép cộng \( + \) 
Phép chia \( / \) 
Phép nhân \( * \) 
Kết quả làm tròn đến chữ số thập phân thứ hai. 
Đầu vào: Hai số thực và phép toán cần thực hiện (trong số các phép toán \(+, -, *, /\)). 
Đầu ra: Kết quả phép toán. 

```python
def may_tinh(a, b, op):
    if op == "+":
        return a + b
    if op == "-":
        return a - b
    if op == "*":
        return a * b
    if op == "/":
        if b == 0:
            return None
        return a / b
a, b, op = input().split()
a = float(a)
b = float(b)
res = may_tinh(a, b, op)
if res == None:
    print("None")
else:
    print(f"{may_tinh(a, b, op):.2f}")
```

# W5A9
Viết hàm trả về tổng các chữ số của số nguyên dương n 

```python
def tong_so_nguyen(n):
    if n <= 0:
        return None
    tong = 0
    while n > 0:
        tong += n % 10
        n //= 10
    return tong
n = int(input())
print(tong_so_nguyen(n))
```

# W5A12
Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và 
0 <= |r| < |divisor|. 

```python
def divide_with_remainder(dividend, divisor):
    if divisor == 0:
        return None, None
    q = dividend // divisor
    r = dividend % divisor
    if r < 0:
        if divisor > 0:
            q -= 1
            r += divisor
        else:
            q += 1
            r -= divisor
    return q, r
```

# W5A15
Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông. 

```python
def is_right_triangle(a, b, c):
    if a + b <= c or a + c <= b or b + c <= a:
        return False
    if a >= b and a >= c:
        huyen, canh1, canh2 = a, b, c
    elif b >= a and b >= c:
        huyen, canh1, canh2 = b, a, c
    else:
        huyen, canh1, canh2 = c, a, b
    return huyen ** 2 == canh1 ** 2 + canh2 ** 2
a, b, c = map(int, input().split())
print(is_right_triangle(a, b, c))
```
